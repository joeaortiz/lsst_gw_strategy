```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

x = np.arange(4,300)
y = 15/np.sqrt(x)

fig = plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.xscale('log')
#plt.yscale('log')
plt.xlabel('Number of joint kilonova detections', size=14)
plt.ylabel('Percentage uncertainty in $H_0$', size=14)
plt.grid( linestyle='--')
plt.tight_layout()
plt.locator_params(numticks=4)

plt.show()

fig.savefig('figures/uncertainty_decay.png')
```
